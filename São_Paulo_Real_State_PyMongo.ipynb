### Database db_imoveis_sp

```python
#Switching to the database db_imoveis_sp
db = client["db_imoveis_sp"]
```

```python
#Selecting the imoveis_sp collection
colecao = db["imoveis_sp"]
```

```python
#Checking the documents in the collection
show_collection(colecao)
```

```python
#Filtering all documents where the neighborhood(bairro) is equal to "paraiso"
resultado = colecao.find({"bairro": "paraiso"})
for documento in resultado:
  print(documento)
```

```python
#Filtering all documents where the neighborhood(bairro) is equal to "Vila-mariana"
resultado = colecao.find({"bairro": "vila-mariana"})
for documento in resultado:
  print(documento)
```

```python
#Filtering all documents with area greater than 60, sorted by area in ascending order, limited to the first 15 documents
resultado = colecao.find({"areaM2": {"$gt": 60}}).sort("areaM2",1).limit(15)
for documento in resultado:
  print(documento)
```

```python
#Filtering all documents with an area equal to 60, sorted by area in ascending order, limited to the first 15 documents
resultado = colecao.find({"areaM2":  60}).sort("areaM2",1).limit(15)
for documento in resultado:
  print(documento)
```

```python
#Filtering all documents with an area greater than or equal to 60, sorted by area in ascending order
resultado = colecao.find({"areaM2": {"$gte": 60}}).sort("areaM2",1)
for documento in resultado:
  print(documento)
```

```python
#filtrando todos os documentos com rea igual a 40 ou 60, ordenados pela rea em ordem crescente
resultado = colecao.find({"areaM2": {"$in": [40, 60]}}).sort("areaM2",1)
for documento in resultado:
  print(documento)
```

```python
#Filtering all documents with an area less than 60, sorted by area in descending order
resultado = colecao.find({"areaM2": {"$lt": 60}}).limit(15).sort("areaM2",-1)
for documento in resultado:
  print(documento)
```

```python
#Filtering all documents with an area less than or equal to 60, sorted by area in ascending order, limited to 20 documents
resultado = colecao.find({"areaM2": {"$lte": 60}}).sort("areaM2",-1).limit(20)
for documento in resultado:
  print(documento)
```

```python
#Filtering all documents with an area not equal to 60, sorted by area in ascending order, and limited to 15 documents
resultado = colecao.find({"areaM2": {"$ne": 60}}).sort("areaM2",1).limit(15)
for documento in resultado:
  print(documento)
```

```python
#Filtering all documents with an area different from 20, 60, and 80, sorted by area in ascending order, and limited to 25 documents
resultado = colecao.find({"areaM2": {"$in": [20, 60, 80]}}).sort("areaM2",1).limit(25)
for documento in resultado:
  print(documento)
```

```python
#Filtering all documents with $or condition, sorted in descending order
resultado = colecao.find({"$or": [{"vagas": 2}, {"suites": 2}]}, {"_id": 0}).sort("vagas",-1).limit(10)
for documento in resultado:
  print(documento)
```

```python
#Filtering all documents with $and condition, sorted in descending order
resultado = colecao.find({"$and": [{"vagas": 2}, {"suites": 2}]}, {"_id": 0}).sort("vagas",-1).limit(10)
for documento in resultado:
  print(documento)
```
